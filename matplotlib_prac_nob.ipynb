```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
##Line Plot
x = np.linspace(0, 10, 100)    #generating 100 values between 0 and 10
y = np.sin(x)    #apply the sin function to every value in the 'x' array

plt.plot(x, y, label = 'Sine Wave', color = 'green', linestyle = '-', linewidth = 2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot: Sine Wave')
plt.legend()
```

```python
##Scatter Plot
x = np.random.rand(100)    #generates 100 random values between 0 and 1
y = 2 * x + np.random.randn(100)    #generates 100 random values from normal distribution 
                                    #creates a straight line(y = 2x) with some noise(2x + randn(100))
plt.scatter(x, y, color = 'blue', label = 'Random Data', marker = 'o')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot: Random Data')
plt.legend()
```

```python
##Bar Chart
cat = ['Cat A', 'Cat B', 'Cat C']
values = [25, 40, 15]

plt.bar(cat, values, label = 'Bar Chart', color = 'red', edgecolor = 'black')
plt.title('Bar Chart: Category Comparision')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
```
