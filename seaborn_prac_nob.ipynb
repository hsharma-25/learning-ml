```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
fmri = sns.load_dataset('fmri')
sns.relplot(
    data = fmri, kind = 'line',
    x = 'timepoint', y = 'signal', col = 'region',
    hue = 'event', style = 'event'
)
```

```python
tips = sns.load_dataset('tips')
print(tips.shape)
tips.info()
tips.tail()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
plt.title('Total bill vs. tip')
plt.show()
```

```python
sns.lineplot(x = 'day', y = 'total_bill', data = tips, style = 'sex', hue = 'sex')
```

```python
sns.histplot(tips['total_bill'], bins = 20, kde = True)    #bins = 20 splits range into 20 equal intervals
#kde(kernel density estimate) gives a smooth version of the histogram
```

```python
iris = sns.load_dataset('iris')
print(iris.shape)
iris.info()
iris.head()
```

```python
iris['species'].value_counts()
```

```python
##Heatmap
#calc. correlation matrix for numeric columns only
corr_mat = iris.select_dtypes(include = 'number').corr()
print(corr_mat)
```

```python
sns.heatmap(corr_mat, annot = True, cmap = 'coolwarm', fmt = '.2f', linewidth = 0.5)
```
